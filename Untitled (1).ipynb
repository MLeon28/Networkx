# Creando un grafo

Por definición, un gráfico es una colección de nodos (vértices) junto con pares de nodos identificados (llamados aristas, enlaces, etc.).
 En NetworkX, los nodos pueden ser cualquier objeto modificable, por ejemplo, una cadena de texto, una imagen, un objeto XML, otro gráfico,
  un objeto de nodo personalizado, etc.

```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
```

**Nodos**

La gráfica G puede crecer de varias maneras. NetworkX incluye muchas funciones de generación de gráficos y facilidades para leer y escribir gráficos en muchos formatos. Sin embargo, para comenzar, veremos manipulaciones simples. Puede agregar un nodo a la vez,


```python
G.add_node(1)
```

o agregue nodos desde cualquier contenedor iterable, como una lista

```python
G.add_nodes_from([2, 3])
```

También puede agregar nodos junto con atributos de nodos si su contenedor produce 2 tuplas de la forma (nodo, nodo_atributo_dict):

```python
G.add_nodes_from([

    (4, {"color": "red"}),

    (5, {"color": "green"}),

])
```

Los atributos de los nodos se analizan más adelante.

Los nodos de un gráfico se pueden incorporar a otro:

```python
H = nx.path_graph(10)

G.add_nodes_from(H)
```

G ahora contiene los nodos de H como nodos de G. Por el contrario, podría usar el gráfico H como un nodo en G.



```python
G.add_node(H)
```

 **Creando un grafo simple**

```python
G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)

# explicitly set positions
pos = {1: (0, 0), 2: (-1, 0.3), 3: (2, 0.17), 4: (4, 0.255), 5: (5, 0.03)}

options = {
    "font_size": 36,
    "node_size": 3000,
    "node_color": "white",
    "edgecolors": "black",
    "linewidths": 5,
    "width": 5,
}
nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

analizando la estructura de la red

```python

A= list(nx.connected_components(G))
print("Los elementos de el grafo son : ", A)

print("Los grados de cada nodo son : ",  G.degree())

print("El clustering por nodo es : ", nx.clustering(G))

print("El promedio del clustering es:", (nx.average_clustering(G)))

print("El promedio del camino más corto es : ", nx.average_shortest_path_length(G))

print("El diametro es: ", nx.diameter(G))

```

**Creando un grafo dirigido**

```python
G = nx.DiGraph([(0, 3), (1, 3), (2, 4), (3, 5), (3, 6), (4, 6), (5, 6)])

# group nodes by column
left_nodes = [0, 1, 2]
middle_nodes = [3, 4]
right_nodes = [5, 6]

# set the position according to column (x-coord)
pos = {n: (0, i) for i, n in enumerate(left_nodes)}
pos.update({n: (1, i + 0.5) for i, n in enumerate(middle_nodes)})
pos.update({n: (2, i + 0.5) for i, n in enumerate(right_nodes)})

nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

analizando la estructura de la red

```python

print("Los grados de cada nodo son : ",  G.degree())


print("El clustering por nodo es : ", nx.clustering(G))

print("El promedio del clustering es:", (nx.average_clustering(G)))

print("El promedio del camino más corto es : ", nx.average_shortest_path_length(G))

```

**Usando un generador de gráficos estocásticos, por ejemplo,**



1. erdos_renyi_graph(n, p[, seed, directed]) :


2. watts_strogatz_graph(n, k, p[, seed])

3. barabasi_albert_graph(n, m[, seed, ...])

4. random_lobster(n, p1, p2[, seed])




```python
# erdos_renyi_graph(n, p[, seed, directed]) :

er = nx.erdos_renyi_graph(20, 0.45)
nx.draw_networkx(er)
#devuelve un gráfico aleatorio, también conocido como gráfico de Erdős-Rényi o gráfico binomial.

#print("Los grados de cada nodo son : ",  er.degree())

#print("El clustering por nodo es : ", nx.clustering(er))

print("El promedio del clustering es:", (nx.average_clustering(er)))

print("El promedio del camino más corto es : ", nx.average_shortest_path_length(er))
```

```python
ws = nx.watts_strogatz_graph(10, 4,0.1)
nx.draw_networkx(ws)
#Devuelve un gráfico de mundo pequeño de Watts-Strogatz.       
print("El promedio del clustering es:", (nx.average_clustering(ws)))

print("El promedio del camino más corto es : ", nx.average_shortest_path_length(ws))
```

```python
ba = nx.barabasi_albert_graph(20, 5)
nx.draw_networkx(ba)
#Devuelve un gráfico aleatorio Barabási–Albert
print("El promedio del clustering es:", (nx.average_clustering(ba)))

print("El promedio del camino más corto es : ", nx.average_shortest_path_length(ba))
```
